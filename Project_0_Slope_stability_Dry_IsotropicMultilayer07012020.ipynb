```python
"""
this is a software designed to perform slope stability analysis, It is developed by
Sena, Hafiz, Jeffery, Cami, Daniel Boateng and Atobra.
"""

# Define user inputs
# user inputs include: slope height, slope angle, soil types and soil properties

# software ouptut include: Factor of safety for the slope

# analysis is done using the modified Bishop method
from math import *

#I think it would be a good idea to create a class for the individual soil/layer. I'm not sure how helpful it would be aside
#modulation and abstraction. I think we can also define some default soil types in this
class layer:
    def __init__(self, layer= 'this layer'):
        self.gamma = float(input('input the unit_weight of the soil in ' + layer + ' (must be a number): '))
        self.phi = float(input('input the internal angle of friction of the soil ' + layer + ' in degrees (must be a number): '))
        self.cohesion =  float(input('input the cohesion of the soil ' + layer + ' (must be a number): '))
        self.thickness =  float(input('input the thickness of the soil ' + layer + ' in meters (must be a number): '))
    
    def __str__(self):
        #this is what would be called if print(layer) is executed
        return [self.gamma, self.phi, self.cohesion]
    
    
#Another object for the slope with all the properties that a slope should have and methods for calculations that can be applied
#the slope. I thought it would be a good idea to have the slope be initiated by either using the layers as inputs or having the
#class prompt user for input if of the layer specifics is no input is provided. I implemented the latter here but the other is
#still something to think about
class Slope:
    def __init__(self):
        self.layers = int(input('How many layers does the slope cut have: '))

        layer_dict = {}
        for a in range(self.layers):
            layer_dict['layer ' + str(a+1)] = layer('layer' + str(a+1))
        self.Soil=layer_dict
        
        self.slope_angle = float(input('input the angle of the slope in degrees(must be a number): '))
        while type(self.slope_angle) != float or self.slope_angle > 90 or self.slope_angle < 0:
            print('Invalid slope angle. Must be a number between zero(0) and ninety(90)')
            self.slope_angle = float(input('input the angle of the slope in degrees(must be a number): '))

            
            
###I dont know how to calculate individual slope radii for unique layers, and is it right to calculate individual slope radii
##for each layer?
###Moreover, I calculated driving and resisting force for each layer in the slope and used their sum as the total driving
###and resisting force of the slope. Is that right?

    def slope_radius(self):
        for i in range(self.layers):
            slope_radii=self.Soil['layer ' + str(i+1)].thickness *sqrt(1+((tan(self.slope_angle)*tan(self.slope_angle))))
        return slope_radii
    
    
    def driving_force(self):
        driving_forces=[]
        for i in range(self.layers):
            driving_forces.append(self.Soil['layer ' + str(i+1)].gamma 
                                  *((180-self.slope_angle)/360)*(pi*(self.slope_radius())**2))
            
        return sum(driving_forces)

           
    def resisting_force(self):
        resisting_forces=[] 
        for i in range(self.layers):
            resisting_forces.append(self.Soil['layer ' + str(i+1)].cohesion 
                                    *((180-self.slope_angle)/360)*(2*pi*self.slope_radius()))
            
        return sum(resisting_forces)

    
    def factor_of_safety(self):
        return self.resisting_force()/self.driving_force()
        

        

Slope1 = Slope()

print('Radius of failure envelope: ',Slope1.slope_radius())


#DF = radius.driving_force(unit_weight,slope_angle,r)
DF = Slope1.driving_force()
print('Destabilizing force:',DF)


# SF is the stabilizing force
#SF = resisting_force(cohesion,slope_angle,r)
SF = Slope1.resisting_force()
print('Stabilizing force:',SF)

FS = Slope1.factor_of_safety()
print('Factor of safety:', FS)

if FS<1:
    print('Slope will fail')
elif FS==1:
    print('Slope is in limiting equilibrium')

else:
    print('Slope is safe')


```
